# Filter data on travel_type_id = 2

```python
filtered_df.info()
```

```python
cleaned_data = filtered_df.drop(['id','user_id','vehicle_model_id','package_id','travel_type_id','from_city_id','to_city_id','Cost_of_error','to_date'],axis=1)

cleaned_data.info()
```

# Transformation

## Transform from_area_id to bins of High Med low

```python
# cleaned_data['from_area_id'] = cleaned_data['from_area_id'].astype('str')  # since IDs present converting them to str

# total_cancel_fai = cleaned_data.groupby('from_area_id')['Car_Cancellation'].sum()
# total_booking_fai = cleaned_data.groupby('from_area_id')['Car_Cancellation'].count()

cleaned_data['cancellation_perc_from_area'] = round(cleaned_data.groupby('from_area_id')['Car_Cancellation'].sum()*100/cleaned_data.groupby('from_area_id')['Car_Cancellation'].count(),2)
cleaned_data['cancellation_perc_from_area'].replace(np.nan,0,inplace=True)
```

```python
cleaned_data.head(100)
```

```python
cancl_density_from_area_conds = [(cleaned_data['cancellation_perc_from_area'].between(0.0,33)),
       (cleaned_data['cancellation_perc_from_area'].between(33,66)),
       (cleaned_data['cancellation_perc_from_area'].between(66,100))]

cancl_density_from_area_labels = ['Low','Medium','High']


cleaned_data['cancl_density_from_area'] = np.select(cancl_density_from_area_conds,cancl_density_from_area_labels)

cleaned_data.head(1000)
```

```python
cancl_density_from_area_conds = [(cleaned_data['cancellation_perc_from_area'].between(0.0,0)),
                                 (cleaned_data['cancellation_perc_from_area'].between(0.01,7.7)),
                                   (cleaned_data['cancellation_perc_from_area'].between(7.7,13.7)),
                                   (cleaned_data['cancellation_perc_from_area'].between(13.7,100))]

cancl_density_from_area_labels = ['No_Cancl','Low','Medium','High']


cleaned_data['cancl_density_from_area'] = np.select(cancl_density_from_area_conds,cancl_density_from_area_labels)

cleaned_data.head(1000)
```
